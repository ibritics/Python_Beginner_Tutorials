##Credit Data
The objective of this task is to assess a dataset containing details from 400 credit card holders and understand the factors that impact a cardholder's Credit Card Balance. The ultimate aim is to predict the average Balance for an individual. This kind of analysis can be a valuable component of customer assessment within a credit card company. The findings from this analysis can help identify customers who may pose a credit default risk and provide insights into the expected behavior of potential customers. Moreover, merging the Credit Card Balance data with information like the credit Limit can aid in calculating a card's credit utilization, a crucial factor in determining a cardholder's credit Rating.

Credit card Balance refers to the average monthly balance across all of the cards owned by a cardholder.
This assumption was made as a result of the Cards variable which refers to the number of credit cards owned by a person and has only one associated Balance figure.
The Balance is calculated as the highest amount incurred on a credit card in a given month. For example if a cardholder spends 400 euro ,  euro 500, and  euro 600 over the course of three months, and each month pays the balance in full, the average balance will be recorded as  euro 500 (i.e. any preliminary balances before the maximum are not taken into account, neither is the final balance of zero).
