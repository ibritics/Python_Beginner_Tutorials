Creating the logic. You should have a flag which indicates whether you have achieved a result or not.

```python
city_to_check = "chicago"
found = 0

########### FOR LOOP############
for city in cities:
    if city == city_to_check.upper():
        found = 1
########## END OF FOR LOOP #######

if found==1:
    print(f"\n{city_to_check} is in the list.")
else:
    print(f"\n{city_to_check} is not in the list.")
```

```python
guests=['aunt', 'sister', 'grandpa', 'dad', 'brother', 'grandma']
```

```python
list_not_invited = [] #New list of not invited
invited_people = ['aunt', 'sister', 'grandpa']
for i in invited_people:
  # print(i)
  # print(guests[-3:])
  # print(guests[3:])

  for n in guests: #nested for loop
    if n not in invited_people and n not in list_not_invited: #not in
      list_not_invited.append(n) #People not in list
  print(f'Hi {i}, unfortunately {",".join(list_not_invited)} will not be for dinner tonight')
```

### Return VS Print

Return returns you a value usefull for further uses in your program

```python
def summar(x):
  return x**2
```

```python
numero = summar(3)
```

```python
print(numero)
```

```python
def add(a, b):
    result = a + b
    return result

return_result = add(3, 5)
```

```python
return_result
```

Print gives you a snapshot of that moment only and not suitable for further uses.

```python
def add(a, b):
    result = a + b
    print(result)

print_result = add(3, 5)
```

```python
print_result
```
